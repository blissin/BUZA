```python
T = int(input())
for test_case in range(1,T+1):
    a=int(input())
    hall_way = []
    for i in range(a):
        b=list(map(int,input().split()))
        hall_way.append(b)

    def find_0_loc(hall_way):
        g_pos=[]
        for i in range(a):
            for j in range(a):
                if hall_way[i][j]!=1:
                    g_pos.append(i,j)

        return g_pos
    
    def find_stop(g_pos, hall_way):
        for ii in range(len(g_pos)):
            y,x=g_pos[ii]
            count=0
            for jj in range(x+1,a):
                if hall_way[y][jj]!=0:
                    break
                else:
                    count+=1
            for jj in range(x-1,-1,-1):
                if hall_way[y][jj]!=0:
                    break
                else:
                    count+=1
            
            for jj in range(y+1,a):
                if hall_way[jj][x]!=0:
                    break
                else:
                    count+=1
            for jj in range(y-1,-1,-1):
                if hall_way[jj][x]!=0:
                    break
                else:
                    count+=1
            
            g_pos[ii].append(count+1)
        
        g_surv = sorted(g_pos, key=lambda x : x[2], reversed=True)

        return g_surv
    
    def surveillance(g_surv, hall_way):
        y,x,z = g_surv
        for jj in range(x,a):
            if hall_way[y][jj] == 1:
                break
            else:
                hall_way[y][jj] += 2
        for jj in range(x,-1,-1):
            if hall_way[y][jj] == 1:
                break
            else:
                hall_way[y][jj] += 2
        
        for jj in range(y,a):
            if hall_way[jj][x] == 1:
                break
            else:
                hall_way[jj][x] += 2
        for jj in range(y,-1,-1):
            if hall_way[jj][x] == 1:
                break
            else:
                hall_way[jj][x] += 2
    
        return hall_way

    new_hall_way = hall_way.copy()
    for i in range(3):
        g_pos = find_0_loc(new_hall_way)
        g_surv = find_stop(g_pos, new_hall_way)
        new_hall_way = surveillance(g_surv[0], new_hall_way)
    
    count_0 = 0
    for i in range(a):
        count_0 += new_hall_way[i].count(0)
    
    print(f'#{test_case}', count_0)
```
